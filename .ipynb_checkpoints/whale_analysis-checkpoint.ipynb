## Rolling Statistics

Risk changes over time. Analyze the rolling statistics for Risk and Beta. 

1. Calculate and plot the rolling standard deviation for the S&P 500 using a 21-day window

### Calculate and plot rolling `std` for all portfolios with 21-day window

```python
# Calculate the rolling standard deviation for all portfolios using a 21-day window
rolling_21days_std = whale_algo_sp500_daily_return.rolling(window=21).std()
rolling_21days_std
# Plot the rolling standard deviation
rolling_21days_std.plot(figsize=(20,10), title="Rolling 21-Days Standard Deviation")
```

### Calculate and plot the correlation

```python
# Calculate the correlation
correlation = whale_algo_sp500_daily_return.corr()
correlation
# Display de correlation matrix
sns.heatmap(correlation, vmax=1, vmin=-1)
```

### Calculate and Plot Beta for a chosen portfolio and the S&P 500

```python
# Calculate covariance of a single portfolio
algo1_covariance = whale_algo_sp500_daily_return["Algo 1"].cov(whale_algo_sp500_daily_return["SP500"])
rolling_algo1_covariance = whale_algo_sp500_daily_return["Algo 1"].rolling(window=21).cov(whale_algo_sp500_daily_return["SP500"])
# Calculate variance of S&P 500
sp500_variance = whale_algo_sp500_daily_return["SP500"].var()
rolling_sp500_variance = whale_algo_sp500_daily_return["SP500"].rolling(window=21).var()
# Computing beta
algo1_beta = algo1_covariance / sp500_variance
rolling_algo1_beta = rolling_algo1_covariance / rolling_sp500_variance
# Plot beta trend
rolling_algo1_beta.plot(figsize=(20,10),title="Rolling 21-Day Beta of Algo 1")
```

## Rolling Statistics Challenge: Exponentially Weighted Average 

An alternative way to calculate a rolling window is to take the exponentially weighted moving average. This is like a moving window average, but it assigns greater importance to more recent observations. Try calculating the [`ewm`](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.ewm.html) with a 21-day half-life.

```python
# Use `ewm` to calculate the rolling window
whale_algo_sp500_daily_return.ewm(halflife='21 days', times=whale_algo_sp500_daily_return.index).mean()
```

# Sharpe Ratios
In reality, investment managers and thier institutional investors look at the ratio of return-to-risk, and not just returns alone. After all, if you could invest in one of two portfolios, and each offered the same 10% return, yet one offered lower risk, you'd take that one, right?

### Using the daily returns, calculate and visualize the Sharpe ratios using a bar plot

```python
# Annualized Sharpe Ratios
sharp_ratio = whale_algo_sp500_daily_return.mean() * 252 / (whale_algo_sp500_daily_return.std() * np.sqrt(252))
```

```python
# Visualize the sharpe ratios as a bar plot
sharp_ratio.plot(kind="bar", figsize=(20,10), title="Sharp Ratio")
```

## Choose 3-5 custom stocks with at last 1 year's worth of historic prices and create a DataFrame of the closing prices and dates for each stock.

For this demo solution, we fetch data from three companies listes in the S&P 500 index.

* `GOOG` - [Google, LLC](https://en.wikipedia.org/wiki/Google)

* `AAPL` - [Apple Inc.](https://en.wikipedia.org/wiki/Apple_Inc.)

* `COST` - [Costco Wholesale Corporation](https://en.wikipedia.org/wiki/Costco)

```python
# Reading data from custom stocks.
okta_data = Path("Resources/okta.csv")
pypl_data = Path("Resources/pypl.csv")
spot_data = Path("Resources/spot.csv")
v_data = Path("Resources/v.csv")

okta = pd.read_csv(okta_data)
okta["Date"] = pd.to_datetime(okta["Date"]).dt.date
okta.index = okta["Date"]
okta = okta.drop(columns=["Date"])

pypl = pd.read_csv(pypl_data)
pypl["Date"] = pd.to_datetime(pypl["Date"]).dt.date
pypl.index = pypl["Date"]
pypl = pypl.drop(columns=["Date"])

spot = pd.read_csv(spot_data)
spot["Date"] = pd.to_datetime(spot["Date"]).dt.date
spot.index = spot["Date"]
spot = spot.drop(columns=["Date"])

v = pd.read_csv(v_data)
v["Date"] = pd.to_datetime(v["Date"]).dt.date
v.index = v["Date"]
v = v.drop(columns=["Date"])
```

```python
# Combine all stocks in a single DataFrame
custom = pd.concat([okta, pypl,spot,v], axis = "columns", join = "inner")
# Drop NAs
custom = custom.dropna()
# Display sample data
custom.tail()
```

```python
# Calculate daily returns
custom_daily_return = custom.pct_change()
# Drop NAs
custom_daily_return = custom_daily_return.dropna()
# Display sample data
custom_daily_return.head()
```

## Calculate the weighted returns for the portfolio assuming an equal number of shares for each stock

```python
# Set weights
weights = [1/4, 1/4, 1/4,1/4]

# Calculate portfolio return
weighted_custom_daily_return = custom_daily_return.dot(weights)

# Display sample data
weighted_custom_daily_return.head()
```

## Join your portfolio returns to the DataFrame that contains all of the portfolio returns

```python
# Join your returns DataFrame to the original returns DataFrame
custom_added_return = pd.concat([whale_algo_sp500_daily_return, weighted_custom_daily_return], axis = "columns", join = "inner")
custom_added_return.dropna()
custom_added_return = custom_added_return.rename(columns={0:"Custom"})
custom_added_return.tail()
```

```python
# Only compare dates where return data exists for all the stocks (drop NaNs)
custom_added_return.isnull().sum()
```
